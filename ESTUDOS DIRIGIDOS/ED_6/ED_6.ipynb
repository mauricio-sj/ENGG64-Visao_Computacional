```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import skimage

# Passo 1: Leitura e exibição da imagem em tons de cinza e formato double
flowers = cv.imread("/content/flowers4.png", cv.IMREAD_GRAYSCALE)
flowers = skimage.img_as_float(flowers)
plt.imshow(flowers, cmap="gray")
plt.title("Imagem original em tons de cinza e formato double")
plt.show()

# Passo 2: Geração e exibição do kernel médio 15x15 com volume unitário (soma igual a 1)
kU = np.ones((15, 15)) / 225
plt.imshow(kU, cmap="gray")
plt.title("Kernel médio 15x15 com volume unitário")
plt.show()

# Passo 3: Aplicação do kernel kU à imagem usando a função filter2D e armazenamento da imagem resultante em imU
imU = cv.filter2D(flowers, -1, kU)

# Passo 4: Geração e exibição do filtro Gaussiano com sigma=5 e largura de 17
kG = cv.getGaussianKernel(17, 5)
plt.imshow(kG, cmap="gray")
plt.title("Filtro Gaussiano (sigma=5, size=17)")
plt.show()

# Passo 5: Aplicação do filtro kG à imagem e armazenamento da imagem resultante em imG
imG = cv.filter2D(flowers, -1, kG)

# Passo 6: Exibição das imagens imU e imG
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
axes[0].imshow(imU, cmap="gray")
axes[0].set_title("Imagem com kernel kU aplicado")
axes[1].imshow(imG, cmap="gray")
axes[1].set_title("Imagem com filtro Gaussiano kG aplicado")
plt.show()
```
