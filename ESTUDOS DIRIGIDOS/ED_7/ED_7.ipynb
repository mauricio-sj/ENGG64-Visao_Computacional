1) O objetivo deste estudo é analisar a quantidade R, G e B a imagem possui
através do histograma. A partir da imagem 'flowers9.png':

a. Faça os histogramas referentes aos planos R, G e B da imagem;

b. Plote a imagem RGB e as imagens de cada plano separadamente;

```python
import cv2 as cv
import matplotlib.pyplot as plt

# Carregando a imagem
image = cv.imread('/content/flowers9.png')

# Convertendo de BGR para RGB
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Separando os canais de cor
r_channel = image_rgb[:, :, 0]
g_channel = image_rgb[:, :, 1]
b_channel = image_rgb[:, :, 2]

# Calculando os histogramas para cada canal
hist_r = cv.calcHist([image_rgb], [0], None, [256], [0, 256])
hist_g = cv.calcHist([image_rgb], [1], None, [256], [0, 256])
hist_b = cv.calcHist([image_rgb], [2], None, [256], [0, 256])

# Plotando os histogramas de cada canal
plt.figure(figsize=(8, 6))
plt.plot(hist_r, color='red')
plt.title('Histograma - Canal R')
plt.xlabel('Intensidade')
plt.ylabel('Frequência')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(hist_g, color='green')
plt.title('Histograma - Canal G')
plt.xlabel('Intensidade')
plt.ylabel('Frequência')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(hist_b, color='blue')
plt.title('Histograma - Canal B')
plt.xlabel('Intensidade')
plt.ylabel('Frequência')
plt.show()

# Plotando a imagem original e os canais separados
plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.imshow(image_rgb)
plt.title('Imagem Original')

plt.subplot(2, 2, 2)
plt.imshow(r_channel, cmap='Reds')
plt.title('Canal R')

plt.subplot(2, 2, 3)
plt.imshow(g_channel, cmap='Greens')
plt.title('Canal G')

plt.subplot(2, 2, 4)
plt.imshow(b_channel, cmap='Blues')
plt.title('Canal B')

plt.tight_layout()
plt.show()
```
